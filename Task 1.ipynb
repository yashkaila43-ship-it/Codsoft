```python
import re
import random

# ---------------------------------------------------------------------------
# Knowledge Base: (pattern, list-of-responses)
# ---------------------------------------------------------------------------
RULES = [
    # Greetings
    (r"\b(hi|hello|hey|howdy|hiya)\b",
     ["Hello! How can I help you today?",
      "Hey there! What's on your mind?",
      "Hi! Nice to meet you."]),

    # How are you
    (r"\bhow are you\b|\bhow('re| are) you doing\b",
     ["I'm just a bot, but I'm doing great! How about you?",
      "Feeling electric! What can I do for you?"]),

    # Name
    (r"\bwhat('s| is) your name\b|\bwho are you\b",
     ["I'm CodBot, your friendly AI assistant!",
      "Call me CodBot — built with ❤️ for the CodSoft internship."]),

    # Help
    (r"\bhelp\b|\bwhat can you do\b|\bwhat do you do\b",
     ["I can answer general questions, tell jokes, give the time, and chat with you!",
      "Ask me anything — jokes, time, or just have a chat!"]),

    # Jokes
    (r"\btell me a joke\b|\bjoke\b|\bfunny\b",
     ["Why don't scientists trust atoms? Because they make up everything! 😄",
      "I told my computer I needed a break. Now it won't stop sending me Kit-Kat ads. 😂",
      "Why do programmers prefer dark mode? Because light attracts bugs! 🐛"]),

    # Time
    (r"\bwhat time is it\b|\bcurrent time\b|\btime\b",
     ["__TIME__"]),  # placeholder handled below

    # Date
    (r"\bwhat('s| is) (today'?s? )?date\b|\btoday\b",
     ["__DATE__"]),

    # Weather (static response — no API)
    (r"\bweather\b",
     ["I don't have live weather data, but you can check weather.com or Google! 🌤️"]),

    # AI / Technology
    (r"\bwhat is (ai|artificial intelligence)\b",
     ["AI is the simulation of human intelligence by machines. It includes ML, NLP, computer vision, and more!"]),

    # CodSoft
    (r"\bcodsoft\b",
     ["CodSoft is a community focused on leadership, learning, and student engagement. Great place to grow!"]),

    # Thanks
    (r"\b(thank(s| you)|thx|ty)\b",
     ["You're welcome! 😊", "Happy to help!", "Anytime! 🙌"]),

    # Bye
    (r"\b(bye|goodbye|see you|cya|exit|quit)\b",
     ["Goodbye! Have a great day! 👋",
      "See you later! Take care! 😊"]),

    # Age
    (r"\bhow old are you\b|\bwhat('s| is) your age\b",
     ["I was just created, so I'm basically a newborn in bot years! 👶"]),

    # Favorite
    (r"\bfavorite (color|colour)\b",
     ["I love electric blue — the color of innovation! 💙"]),

    (r"\bfavorite (food|drink)\b",
     ["I run on electricity, but I'd love to try pizza! 🍕"]),

    # Fallback is handled separately
]

# ---------------------------------------------------------------------------
# Chatbot Engine
# ---------------------------------------------------------------------------

def get_time():
    from datetime import datetime
    return f"The current time is {datetime.now().strftime('%I:%M %p')}."

def get_date():
    from datetime import datetime
    return f"Today is {datetime.now().strftime('%A, %B %d, %Y')}."

def get_response(user_input: str) -> str:
    text = user_input.lower().strip()

    for pattern, responses in RULES:
        if re.search(pattern, text):
            reply = random.choice(responses)
            if reply == "__TIME__":
                return get_time()
            if reply == "__DATE__":
                return get_date()
            return reply

    # Fallback
    fallbacks = [
        "Hmm, I'm not sure about that. Could you rephrase?",
        "Interesting! Tell me more.",
        "I didn't quite catch that. Try asking something else!",
        "I'm still learning. Ask me something else! 😊",
    ]
    return random.choice(fallbacks)

# ---------------------------------------------------------------------------
# Main Loop
# ---------------------------------------------------------------------------

def main():
    print("=" * 50)
    print("        CodBot - Rule-Based Chatbot")
    print("   (Type 'bye' or 'quit' to exit)")
    print("=" * 50)

    while True:
        user_input = input("\nYou: ").strip()
        if not user_input:
            continue

        response = get_response(user_input)
        print(f"CodBot: {response}")

        # Exit condition
        if re.search(r"\b(bye|goodbye|exit|quit|cya)\b", user_input.lower()):
            break

if __name__ == "__main__":
    main()
```
